```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\Users\Adeel\Pictures\tree.jpg")
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
img_gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gry,cmap='grey')
plt.axis('off')
plt.show()
```

```python
resize_img=cv2.resize(img,(400,400))
img_rgb=cv2.cvtColor(resize_img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
plt.title('orignal gray_scale image')
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
negative_img=255-img
plt.imshow(negative_img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
_,binary_img=cv2.threshold(img,127,255,cv2.THRESH_BINARY)

plt.imshow(binary_img, cmap='gray')
plt.axis('off')
plt.show()
```
