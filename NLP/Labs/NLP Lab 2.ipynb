# Removing number from string

```python
import re
string ='''	 Python 3.0, released in 2008, was a major revision of the language that is not completely backward compatible and 
much Python 2 code does not run unmodified on Python 3. With Python 2's end-of-life, only Python 3.6.x[30] and 
later are supported, with older versions still supporting e.g. Windows 7 (and old installers not restricted to 64-bit Windows).'''
lower_string = string.lower()
remove_num=re.sub(r'\d+','',lower_string)
print(remove_num)
```

# removing punctuation from string

```python
remove_punctuation=re.sub(r'[^\w\s]','',remove_num)
print(remove_punctuation)
```

# removing extra spaces

```python
remove_wide_space=remove_punctuation.strip()
print(remove_wide_space)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stop_words=set(stopwords.words('english'))
print(stop_words)
```

```python
lst_string=[remove_wide_space][0].split()
print(lst_string)
```

# removing stopwords from string

```python
no_stopword_string=''
for i in lst_string:
    if not i in stop_words:
        no_stopword_string+=i+' '

no_stopword_string=no_stopword_string[:-1]
print(no_stopword_string)
```

# Importing pdf and doing the same thing on pdf

```python
import PyPDF2

file_path = r"C:\Users\Adeel\Desktop\NLP\SE Assugnment 1 (1).pdf"

with open(file_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)
    content = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])
    print(content)
```

```python
lower_string = content.lower()
remove_num=re.sub(r'\d+','',lower_string)
print(remove_num)
```

```python
remove_punct=re.sub(r'[^\w\s]','',lower_string)
print(remove_punct)
```

```python
no_wide_space=remove_punct.strip()
print(no_wide_space)
```
